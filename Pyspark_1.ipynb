```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Read CSV Example") \
    .getOrCreate()

# Read CSV file
df = spark.read.format("csv") \
    .option("inferSchema", "true") \
    .option("header", "true") \
    .load("/content/atlantic.csv")

# Show data
df.show()

# Check schema
df.printSchema()
```
