```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ColabSparkSession").getOrCreate()
```

```python
emp_data = [
(1,'manish',26,20000,'india','IT'),
(2,'rahul',None,40000,'germany','engineering'),
(3,'pawan',12,60000,'india','sales'),
(4,'roshini',44,None,'uk','engineering'),
(5,'raushan',35,70000,'india','sales'),
(6,None,29,200000,'uk','IT'),
(7,'adam',37,65000,'us','IT'),
(8,'chris',16,40000,'us','sales'),
(None,None,None,None,None,None),
(7,'adam',37,65000,'us','IT')
]
my_schema=['id','name','age','salary','country','dept']
df=spark.createDataFrame(data=emp_data,schema=my_schema)
```

```python
df.count()
```

```python
from pyspark.sql.functions import count
df.select(count('salary')).show()
##Discards null value
```

All Aggregation Functions

```python
from pyspark.sql.functions import min, max, avg
df.select(min('salary').alias('total sal'), max('salary').alias('max_sal'), avg('salary').alias('avg_sal'),count('salary').alias('cnt')).show()
```
