##Printing Bad Records-

Mode

Permissive-all records will be shown

Dropmalformed- Display bad record as null

FailFast - Show error

For display bad record create my_schema

```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Read CSV Example") \
    .getOrCreate()

# Read CSV file
df = spark.read.format("csv") \
    .option("header", "true") \
    .schema(my_schema) \
    .option("mode","PERMISSIVE") \
    .load("/content/badrecord.xlsx")


# Show data
df.show()

# Check schema
df.printSchema()
```

```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Read CSV Example") \
    .getOrCreate()

# Read CSV file
df = spark.read.format("csv") \
    .option("header", "true") \
    .schema(my_schema) \
    .option("mode","FAILFAST") \
    .load("/content/badrecord.xlsx")


# Show data
df.show()

# Check schema
df.printSchema()
```

```python
my_schema=StructType([
    StructField("DEST_COUNTRY_NAME",StringType(),True),
    StructField("ORIGIN_COUNTRY_NAME",StringType(),True),
    StructField("count",IntegerType(),True)
])
```

```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Read CSV Example") \
    .getOrCreate()

# Read CSV file
df = spark.read.format("csv") \
    .option("header", "true") \
    .schema(my_schema) \
    .option("badRecordsPath","/content/badrecord.xlsx") \
    .load("/content/badrecord.xlsx")


# Show data
df.show(truncate=False)

# Check schema
df.printSchema()
```
