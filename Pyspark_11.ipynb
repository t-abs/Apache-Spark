Repartition and Coalesce

```python
from pyspark.sql import SparkSession
```

```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Read CSV Example") \
    .getOrCreate()

```

```python
df=spark.read.format("csv") \
   .option("header", "true") \
   .option("inferSchema","true") \
   .load("/content/flight_data.csv")
```

```python
df=df.rdd.getNumPartitions()
```

```python
df=spark.read.format("csv") \
   .option("header", "true") \
   .option("inferSchema","true") \
   .load("/content/flight_data.csv")
partitioned_flight=df.repartition(4)
```

```python
grouped_multi_df = partitioned_flight.groupBy(
    "DEST_COUNTRY_NAME",
    "ORIGIN_COUNTRY_NAME"
).sum("count")

grouped_multi_df.show()
```

```python
from pyspark.sql.functions import sum

result_df = partitioned_flight.groupBy("DEST_COUNTRY_NAME") \
    .agg(sum("count").alias("total_flights"))

result_df.show()
```

```python
coalesced_flight = df.coalesce(1)
```

```python
coalesced_flight.rdd.getNumPartitions()
```

```python
from pyspark.sql.functions import sum

grouped_df = coalesced_flight.groupBy("DEST_COUNTRY_NAME") \
    .agg(sum("count").alias("total_flights"))

grouped_df.show()
```

```python
grouped_multi_df = coalesced_flight.groupBy(
    "DEST_COUNTRY_NAME",
    "ORIGIN_COUNTRY_NAME"
).agg(sum("count").alias("total_flights"))

grouped_multi_df.show()
```
