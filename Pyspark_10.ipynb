Union VS UnionAll

```python
from pyspark.sql import SparkSession
```

```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Read CSV Example") \
    .getOrCreate()

```

```python
data=[(10 ,'Anil',50000, 18),
(11 ,'Vikas',75000,  16),
(12 ,'Nisha',40000,  18),
(13 ,'Nidhi',60000,  17),
(14 ,'Priya',80000,  18),
(15 ,'Mohit',45000,  18),
(16 ,'Rajesh',90000, 10),
(17 ,'Raman',55000, 16),
(18 ,'Sam',65000,   17)]
schema=['id','Name','Salary','Age']
manager_df=spark.createDataFrame(data=data,schema=schema)
manager_df.show()
```

```python
data1=[(19 ,'Sohan',50000, 18),
(20 ,'Sima',75000,  17)]
schema=['id','Name','Salary','Age']
manager_df1=spark.createDataFrame(data=data1,schema=schema)
manager_df1.show()
```

```python
manager_df.union(manager_df1).show()
```

```python
manager_df.union(manager_df1).count()
```

```python
manager_df.unionAll(manager_df1).count()
```

```python
manager_df1.createOrReplaceGlobalTempView("manager_dftbl")
manager_dp.createOrReplaceGlobalTempView("manager_dptbl")
```

```python
spark.sql(
    "select * from global_temp.manager_dftbl UNION select * from global_temp.manager_dptbl"
).count()
```

```python
spark.sql(
    "select * from global_temp.manager_dftbl UNION ALL select * from global_temp.manager_dptbl"
).count()
```

Note-

In Data Frame there is no difference in union and union all

whearas in spark SQL Union All contains all duplicate and Union drops duplicate records

```python
wrong_column_data = [
    (19, 50000,  'Sohan'),
    (20, 75000, 'Sima')
]

schema = ['id', 'Salary', 'Name']

wrong_df = spark.createDataFrame(wrong_column_data, schema=schema)
wrong_df.show()
```

```python
correct_data = [
    (19, 'Sohan', 18, 50000),
    (20, 'Sima', 17, 75000)
]

correct_schema = ['id', 'Name', 'Age', 'Salary']

correct_df = spark.createDataFrame(correct_data, correct_schema)
correct_df.show()
```

```python
wrong_df.union(correct_df).show()
```
