```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Read CSV Example") \
    .getOrCreate()

```

```python
emp_data = [
(1,'manish',26,20000,'india','IT'),
(2,'rahul',None,40000,'germany','engineering'),
(3,'pawan',12,60000,'india','sales'),
(4,'roshini',44,None,'uk','engineering'),
(5,'raushan',35,70000,'india','sales'),
(6,None,29,200000,'uk','IT'),
(7,'adam',37,65000,'us','IT'),
(8,'chris',16,40000,'us','sales'),
(None,None,None,None,None,None),
(7,'adam',37,65000,'us','IT')
]
emp_columns = ["emp_id","emp_name","emp_age","emp_salary","emp_country","emp_dept"]
df = spark.createDataFrame(data=emp_data, schema = emp_columns)
df
```

```python
from pyspark.sql.functions import col, when

df.withColumn("adult", when(col("emp_age") > 18, "Yes")
                        .when(col("emp_age") < 18, "No")
                         .otherwise("No Value")).show()
```

```python
df.createOrReplaceTempView("employee_data")
```

```python
spark.sql("""
    SELECT
        emp_id,
        emp_name,
        emp_age,
        emp_salary,
        emp_country,
        emp_dept,
        CASE
            WHEN emp_age > 18 THEN 'Yes'
            WHEN emp_age < 18 THEN 'No'
            ELSE 'No Value'
        END AS adult
    FROM employee_data
""").show()
```
