```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ColabSparkSession").getOrCreate()
```

Write DataFrame in Disk



```python
df=spark.read.format("csv")\
    .option("header","true")\
    .option("inferSchema","true")\
    .load("/content/country_comparison_large_dataset.csv")
df.show()
```

```python
df.write \
  .partitionBy("country", "year") \
  .mode("overwrite") \
  .saveAsTable("demo_db.country_year_partitioned")
df.show()
```

 DataFrame write (saving to disk)

Purpose:
Save in-memory Spark DataFrame to permanent storage (HDFS / S3 / ADLS / local / Hive).

Syntax:

df.write.format("parquet").mode("overwrite").save(path)


Why needed:
DataFrame is temporary  .write() makes it persistent.

Common formats: parquet , csv, json, orc
Modes: overwrite, append, ignore, error

 partitionBy

What it does:
Physically splits data into folders based on column values.

Purpose:
Speed up filtering (WHERE clause) using partition pruning.

Example use:

.partitionBy("country", "year")


Best for:

Low-cardinality columns

Frequently filtered columns

Examples: country, year, date

Avoid for:
High-cardinality columns (user_id, txn_id)

 bucketBy

What it does:
Splits data into a fixed number of files using hash of a column.

Purpose:
Speed up joins & aggregations by reducing shuffle.

Example use:

.bucketBy(8, "customer_id")


Best for:

High-cardinality join columns

Repeated joins

Important rule:
 Works only with saveAsTable()
 Cannot bucket and partition on the same column
