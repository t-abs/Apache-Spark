```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Read CSV Example") \
    .getOrCreate()

```

```python

data=[(10 ,'Anil',50000, 18),
(11 ,'Vikas',75000,  16),
(12 ,'Nisha',40000,  18),
(13 ,'Nidhi',60000,  17),
(14 ,'Priya',80000,  18),
(15 ,'Mohit',45000,  18),
(16 ,'Rajesh',90000, 10),
(17 ,'Raman',55000, 16),
(18 ,'Sam',65000,   17),
(15 ,'Mohit',45000,  18),
(13 ,'Nidhi',60000,  17),
(14 ,'Priya',90000,  18),
(18 ,'Sam',65000,   17)
     ]
emp_data=spark.createDataFrame(data,['emp_id','emp_name','emp_salary','emp_age'])
```

```python
emp_data.distinct().show()
```

```python
emp_data.select("emp_id","emp_name").distinct().show()
```

```python
emp_data.dropDuplicates().show()
```

```python
emp_data.dropDuplicates(['emp_name','emp_id']).show()
```

```python
emp_data.sort(col("emp_salary")).show()
```

```python
from pyspark.sql.functions import *
```

```python
emp_data.sort(col("emp_salary").desc()).show()
```

```python
emp_data.sort(col("emp_salary").asc()).show()
```
