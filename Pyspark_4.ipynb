How to read json file in json?

Types of json-

1. Line-delimited json

2. Multilined json

3. Nested json

```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder \
    .appName("Read CSV Example") \
    .getOrCreate()

# Read CSV file
df = spark.read.format("json") \
    .option("inferSchema", "true") \
    .option("header", "true") \
    .load("/content/line-delimited.json")

# Show data
df.show()

# Check schema
df.printSchema()
```

```python
df1=spark.read.format("json") \
    .option("inferSchema","true") \
    .option("multiline","true") \
    .load("/content/Multilined.json")
df1.show()
df1.printSchema()
```

```python
df1=spark.read.format("json") \
    .option("inferSchema","true") \
    .option("multiline","true") \
    .load("/content/Nested.json")
df1.show()
df1.printSchema()
```
