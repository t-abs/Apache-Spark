```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ColabSparkSession").getOrCreate()
```

How to create Dataframe in Spark-

A Spark DataFrame is a distributed collection of data organized into named columns, conceptually equivalent to a table in a relational database or a data frame in R/Python, but with richer optimizations for big data processing.




```python
my_data = [
(1 , 1,),
(2 , 3,),
(3, 4,),
]
```

```python
my_schema=['id','value']
```

```python
spark.createDataFrame(data=my_data,schema=my_schema).show()
```
